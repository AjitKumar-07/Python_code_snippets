### loc vs iloc

The mai diff b/w these two are:-

* loc gets rows (and/or columns) with particular labels.

* iloc gets rows (and/or columns) at integer locations.

```python
s= pd.Series(list('abcde'))
s
```

```python
s1= pd.Series(list('abcde'), index=[49,48,0,1,2])
s1
```

```python
s1.loc[0] # value at index label 0
```

```python
s1.iloc[0] # value at index location 0
```

```python
s1.loc[0:1] # rows at index labels b/w 0 and 1 (inclusive)
```

```python
s1.iloc[0:1] # rows at index location b/w 0 and 1 (exclusive)
```

```python
s1.iloc[0:2]
```

### what if you sort the series

```python
s1 =s1.sort_index()
s1
```

```python
s1.loc[0]  # value at index label 0
```

```python
s1.iloc[0]  # value at index location 0
```

```python
s1.loc[0:1]  # rows at index labels b/w 0 and 1 (inclusive)
```

```python
s1.iloc[0:1]  # rows at index location b/w 0 and 1 (exclusive)
```
